# File Handling

## File :-

#### File is the collection of data is available to a program. We can retrieve and use data stored in a file whenever, we required.
#### e.g image,text,doc,pdf,etc



### Advantages :-

#### 1- Stored data is permanent unless someone remove it.
#### 2- Stored data can be shared.
#### 3- It is possible to update or remove the data.

### Types of Files :-

#### 1- Text file :- It stores the data in the form of characters. It is used to store characters and strings.

#### 2- Binary File :- It stores data in the form of bytes, a group of 8 bits each. It is used to store text, images, pdf, csv, video and audio.

## Text Mode and Binary Mode


#### Text Mode - A file opened in text mode, treats its contents as if it contains text strings of the str type. When you get data from a text mode file, Python first decodes the raw bytes using either platform-dependent encoding or specified one.


#### Binary Mode- A file opened in binary mode, Python uses the data in the file without any decoding, binary mode file reflects the raw data in the file.


```python
f= open('student.txt', mode='w')
f.write('Hello\n')
f.write('All\n')
f.write('How are you?\n')
f.close()
print("Writing Success")


f= open('student.txt', mode='r')  
data= f.read()
print(data)
f.close()


f= open('student.txt', mode='rb')
data= f.read()
print(data)
f.close()
```

```python
with open('student.txt', mode='w') as f:
    f.write('Hello\n')
    f.write('All\n')
    f.write('How are you?\n')
print("Writing Success")


with open('student.txt', mode='r') as f: 
    data= f.read()
print(data)



with open('student.txt', mode='rb') as f: 
    data= f.read()
print(data)
```

### Opening a File

#### If you want to use a file or its data, first we have to open it. "open( )" function is used to open a file. 

#### Syntax:- open("file_name", mode= "r"/ "w" /"a" /"r+")

#### "rb" = read
####  "wb" = write
####  "ab" = append
####  "xb" = new file creation with write
####  "rb+" = read and then write
####  "wb+" = write and then read
####  "ab+" = append and then read

```python
## Read

f =open("student.txt", "r")
data=f.read()
print(data)
f.close()


with open("student.txt", "r") as f:
    data= f.read()
    print(data)
          
```

```python
## Write

f = open("student1.txt", "w")
f.write("Python\n")
f.write("Language\n")
f.close()
```

```python
## Append

f = open("student1.txt", "a")
f.write("Is\n")
f.write("Awesome")
f.close()
```

```python
## X

f = open("Python1.txt", "x")
f.write("Hiiiiii")
f.close()

```

```python
## Read / Write

f =open("read.txt", "r+")

f.write("How are you\n")
f.write("Good Morning\n")
f.close()
```
